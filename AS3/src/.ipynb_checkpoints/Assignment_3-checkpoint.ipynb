```python
import math
import numpy as np
import matplotlib.pyplot as plt
def generate_line_points(angle,d,num_points=25):

    angle=math.radians(angle)
    x=d*math.cos(angle)
    y=d*math.sin(angle)
#     print(x,y)
    xn = np.linspace(0, x, num_points)
    yn = np.linspace(0, y, num_points)
#     print('\n',x_vals,y_vals)
    points=list(zip(xn, yn))
    return points,xn,yn
fig, ax = plt.subplots()
x=[]
y=[]
for angle_deg in range(0, 90, 15):
    points,xn,yn = generate_line_points(angle_deg, 1, num_points=25)
    x.append(xn)
    y.append(yn)
    ax.scatter(*zip(*points), label=f'{angle_deg} degrees')

xs=x.copy()
ys=y.copy()
ax.legend()
plt.show()
# print(xn)
```

```python
datagen = ImageDataGenerator()
train_generator = datagen.flow_from_directory(
    directory=train_dir,
    target_size=(128, 128),
    color_mode="rgb",
    batch_size=32,
    class_mode="binary",
    shuffle=True,
    seed=42
)
test_generator = datagen.flow_from_directory(
    directory=test_dir,
    target_size=(128, 128),
    color_mode="rgb",
    batch_size=32,
    class_mode="binary",
    shuffle=True,
    seed=42
)
validation_generator = datagen.flow_from_directory(
    directory=validation_dir,
    target_size=(128, 128),
    color_mode="rgb",
    batch_size=10,
    class_mode="binary",
    shuffle=True,
    seed=42
)
```

```python

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle 
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers,Sequential
from keras.layers import Input, Dense, Reshape, Flatten, Conv2DTranspose, concatenate, Activation, LeakyReLU,BatchNormalization,Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam,SGD,Adagrad
from keras.models import Model
from keras.layers import Input,Dropout,add,BatchNormalization
import numpy as np

def build_model():
    inputs = Input((128,128,3))
    conv0=Conv2D(32,(3,3),padding='same',activation='relu')(inputs)
    conv0=BatchNormalization()(conv0)
    conv0=Dropout(0.25)(conv0)
    pool0=MaxPooling2D((2,2))(conv0)
    
    conv1=Conv2D(64,(3,3),padding='same',activation='relu')(pool0)
    conv1=BatchNormalization()(conv1)
    conv1=Dropout(0.25)(conv1)
    pool1=MaxPooling2D((2,2))(conv1)
    
    conv2=Conv2D(128,(3,3),padding='same',activation='relu')(pool1)
    conv2=BatchNormalization()(conv2)
    conv2=Dropout(0.25)(conv2)
    pool2=MaxPooling2D((2,2))(conv2)
    flatten=Flatten()(pool2)
    
    dense=Dense(256,activation='relu')(flatten)
    dense=Dropout(0.5)(dense)
    output=Dense(1,activation='sigmoid')(dense)
    
    
    model=Model(inputs,output)
    model.compile(optimizer=Adam(learning_rate=1e-3,beta_1=0.3),loss='binary_crossentropy',metrics=['accuracy'])
    return model


model=build_model()
model.summary()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=50,
    epochs=30,validation_data=validation_generator,validation_steps=len(validation_generator)//5,verbose=1)
```

```python
from tensorflow.keras.initializers import GlorotNormal,HeNormal,RandomNormal

import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle 
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers,Sequential
from keras.layers import Input, Dense, Reshape, Flatten, Conv2DTranspose, concatenate, Activation, LeakyReLU,BatchNormalization,Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam,SGD,Adagrad
from keras.models import Model
from keras.layers import Input,Dropout,add,BatchNormalization
import numpy as np
```

```python
def inception_block(input,filter_size):
    conv0=Conv2D(filter_size,(1,1),kernel_initializer=HeNormal(),padding='same')(input)
    conv0=LeakyReLU(alpha=0.1)(conv0)
    conv0=BatchNormalization()(conv0)
    
    conv1=Conv2D(filter_size*2,(3,3),kernel_initializer=HeNormal(),padding='same')(input)
    conv1=LeakyReLU(alpha=0.1)(conv1)
    conv1=BatchNormalization()(conv1)
    
    conv2=Conv2D(filter_size*4,(5,5),kernel_initializer=HeNormal(),padding='same')(input)
    conv2=LeakyReLU(alpha=0.1)(conv2)
    conv2=BatchNormalization()(conv2)
    
    pool0=MaxPooling2D((3,3), strides=(1,1),padding='same')(input)
    conv=tf.keras.layers.concatenate([conv0,conv1,conv2,pool0])
    return conv



def build_model():
    inputs = Input((32,32,3))
    conv1=inception_block(inputs,8)
    conv2=inception_block(conv1,8)
    conv3=inception_block(conv2,8)
    flatten=Flatten()(conv3)
    dense=Dense(64,activation='relu')(flatten)
    dense=Dropout(0.5)(dense)
    output=Dense(10,activation='softmax')(dense)
    model=Model(inputs,output)
    return model

model=build_model()
model.compile(optimizer=Adam(learning_rate=1e-3,beta_1=0.3),loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True)
```

```python
history = model.fit(X,y_,batch_size=64,epochs=30,validation_split=0.1,verbose=1)
```

```python
def resnet_block(input,filter_size):
    conv=Conv2D(filter_size,(3,3),kernel_initializer=RandomNormal(),padding='same')(input)
    conv=LeakyReLU(alpha=0.1)(conv)
    conv=BatchNormalization()(conv)
    conv=Conv2D(filter_size,(3,3),kernel_initializer=RandomNormal(),padding='same')(conv)
    conv=LeakyReLU(alpha=0.1)(conv)
    conv=BatchNormalization()(conv)
    conv=tf.keras.layers.concatenate([input,conv])
    return conv




def build_model():
    inputs=Input((32,32,3))
    conv0=Conv2D(16,(3,3),padding='same',name='Convolution_0')(inputs)
    conv0=BatchNormalization()(conv0)
    conv0=LeakyReLU(alpha=0.1)(conv0)
    conv0=resnet_block(conv0,16)
    conv0=Dropout(0.4)(conv0)
    pool0=MaxPooling2D((2,2))(conv0)
    
    conv1=Conv2D(16,(3,3),padding='same',name='Convolution_1')(pool0)
    conv1=BatchNormalization()(conv1)
    conv1=LeakyReLU(alpha=0.1)(conv1)
    conv1=resnet_block(conv1,16)
    conv1=Dropout(0.4)(conv1)
    pool1=MaxPooling2D((2,2))(conv1)
    
    flatten=Flatten()(pool1)
    dense1 = Dense(224, activation='relu')(flatten)
    dense1 = Dropout(0.5)(dense1)
    outputs = Dense(10,activation='softmax')(dense1)
    # kernel_initializer=GlorotNormal()
    model = Model(inputs, outputs)
    return model

model=build_model()
model.compile(optimizer=Adam(learning_rate=1e-3,beta_1=0.3),loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
```
